```python
data['Age'].isnull().sum()
```

```python
data['Age'].fillna(data['Age'].mean(), inplace=True)
```

```python
Age_fig = px.box(data, x='Age')
Age_fig.show()
```

```python
def find_outliers_IQR(df):
      q1=df.quantile(0.25)
      q3=df.quantile(0.75)
      IQR=q3-q1
      outliers = df[((df<(q1-1.5*IQR))|(df>(q3+1.5*IQR)))]
      return outliers
```

```python
Ageoutliers = find_outliers_IQR(data['Age'])
print('No of outliers : ', len(Ageoutliers))
data.drop(Ageoutliers.index, inplace=True)
Ageoutliers = find_outliers_IQR(data['Age'])
print('No of outliers after deleted : ', len(Ageoutliers))
```
